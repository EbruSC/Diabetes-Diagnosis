# Pima Indians Diabetes Data is a simple dataset provided by the UCI Machine Learning dataset, which contains a medical record of Indian patients.

```python
import pandas as pd
#Load Data
data = pd.read_csv('diabetes.csv')
x = data.drop("Outcome", axis=1)
y = data["Outcome"]
```

```python
print(data)
```

```python
print(x)
```

```python
print(y)
```

```python
#Define Keras Model
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(12, input_dim=8, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
#Compile The Keras Model
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
#Start Training (Fit the Model)
history=model.fit(x,y, epochs=350, batch_size=10)
```

```python
#Evaluate the Model
_, accuracy = model.evaluate(x, y)
print("Model accuracy: %.2f"% (accuracy*100))
```

```python
#make predictions
predictions = model.predict(x)
print([round(x[0]) for x in predictions])
```

```python
from matplotlib import pyplot as plt
# Plot training & validation accuracy values
plt.figure(figsize=(14,3))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])

plt.title('Model Doruluk')
plt.ylabel('Doruluk')
plt.xlabel('Adm')
plt.legend(['Train'], loc='upper left')
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])

plt.title('Model Kayp')
plt.ylabel('Kayp')
plt.xlabel('Adm')
plt.legend(['Train'], loc='upper left')
plt.show()
```
